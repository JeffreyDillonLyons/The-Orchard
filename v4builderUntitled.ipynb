```python
import random
import turtle
from turtle import Turtle
import time


def setup():
    turtle.setup(width=1.0,height=1.0)
    turtle.setworldcoordinates(0,0,1000,1000)
    turtle.speed(0)
    turtle.bgcolor('yellow green')
    window = turtle.Screen()
    window.title("Amazing Tree")
#Draw the edges,text, etc of the garden
    turtle.penup()
    turtle.goto(300,750)
    turtle.color('dark red')
    turtle.write("Welcome to the Orchard, young wizard", font=("Helvetica", 30, "italic"))
    turtle.home()
    turtle.pendown
    turtle.color('gold')
    turtle.pensize(25)
    turtle.(900)
    turtle.left(90)
    turtle.forward(990)
    turtle.left(90)
    turtle.forward(990)
    turtle.left(90)
    turtle.home()
setup()


t1 = Turtle()
t2 = Turtle()
t3 = Turtle()
t4 = Turtle()

######## Variable Inputs:
angle = random.randint(10,20)              # Angels of new branches.
tree_age = random.randint(1,2)                               # How old  the tree will be 
branch_length = 35                         # Branch length. For now, one value.
branch_thickness = branch_length/17        # Branch length. For now, one value.
x2 = random.uniform(0.5,0.8)               # Specifcy how much the branches reduce in size. Uniform function: 
                                           # Return a random floating point number N such that a <= N <= b for a <= b and b <= N <= a for b < a.
######## Turtle screen settings
jeff = turtle.Turtle()
jeff.speed(0)                            # using screen.tracer() speeds up your drawing (by skipping some frames when drawing)
#window.tracer(2)


#Turtle settings:
turtles = [t1, t2, t3, t4]
for jeff in turtles:
    
    x_begin = random.uniform(100,700)     # maybe need to use += instead of =  but not sure yet.
    y_begin = random.uniform(200,150)
    jeff.penup()
    jeff.goto(x_begin, y_begin)
    jeff.pendown()
    jeff.dot(30,'tan')
    jeff.left(90)
    
#jeff.setpos(x_put,0)                      # This is where, eventually, the orchard positions new to come in
# Get ready for drawing

jeff.pendown()
jeff.color('peru')
jeff.pensize(6)
stack = []    


############## Turtle's initial drawing, before the tree



# Position settings, draw the ground for the tree to grow in    


                               # List to stack the previous position.

##################### The L-System Set-up:

def apply_rules(letter):                      #Apply rules to an individual letter, and return the result."""
    if letter == 'F':                         
        new_string = 'F[-F][+F]'              # L-system rules
    else:                                     # no rules apply so keep the character
        new_string = letter
    return new_string

def process_string(original_string): #Apply rules to a string, one letter at a time, and return the result."""
    tranformed_string = ""
    for letter in original_string:
        tranformed_string = tranformed_string + apply_rules(letter)
    return tranformed_string

def create_l_system(number_of_iterations, axiom):    #Begin with an axiom, and apply rules to the original axiom string number_of_iterations times, then return the result.
    start_string = axiom
    for counter in range(number_of_iterations):
        end_string = process_string(start_string)
        start_string = end_string
    return end_string

def drawer_Lsystem_tree(treedrawer, instructions, angle, branch_length):    #Draw, using L-system instructions.
    for task in instructions:
        if task == 'F':
            treedrawer.color('peru')
            treedrawer.pensize(branch_length/15)
            treedrawer.forward(branch_length)
            treedrawer.color('green')
            treedrawer.stamp()
            treedrawer.color('peru')
        elif task == '+':
            treedrawer.right(angle)
        elif task == '-':
            treedrawer.left(angle)
        elif task =='[':                              #  [ saves the current position the drawer
            angle2 = treedrawer.heading()
            pos = [treedrawer.xcor(), treedrawer.ycor()]
            stack.append((angle, pos))                # Push state to save
        elif task == ']':                             # Return the drawer to the position set in the last [
            angle, pos = stack.pop()                  # Pop state to restore
            treedrawer.setheading(angle2)
            treedrawer.penup()
            treedrawer.goto(pos[0], pos[1])
            treedrawer.pendown()
        elif task == 'C':
            drawer_Lsystem_tree(treedrawer, instructions, angle, branch_length*x2)
             

############################################################################

# create the string of turtle instructions (in print)
instruction_string = create_l_system(tree_age, "F")
print(instruction_string)

for jeff in turtles:
    drawer_Lsystem_tree(jeff, instruction_string, angle, branch_length)      # draw the picture, using angle 'angle' and segment length 'x'
```
